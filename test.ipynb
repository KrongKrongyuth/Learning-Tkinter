```python
def Grade(score):
    if score >= 80:
        return("A")
    elif 75 <= score < 79:
        return("B+")
    elif 70 <= score < 75:
        return("B")
    elif 65 <= score < 69:
        return("C+")
    elif 60 <= score < 65:
        return("C")
    elif 55 <= score < 59:
        return("D+")
    elif 50 <= score < 55:
        return("D")
    elif score < 50:
        return("F")
```

```python
run_df = pd.DataFrame({"1st Run" : [None, None, None, None, None],
                    "2nd Run" : [None, None, None, None, None],
                    "3rd Run" : [None, None, None, None, None],
                    "Average" : [None, None, None, None, None]})
run_df
```

```python
def cal_time(index, round):
    for i in range(4):
        T_init = time.time()
        for row in range(round):
            assign_score = main_df["Assignment 1"][row] + main_df["Assignment 2"][row] + main_df["Assignment 3"][row] + main_df["Assignment 4"][row] + main_df["Group"][row]
            assign_score = (assign_score / 30) * 100

            midterm_score = (main_df["Midterm"][row] / 35) * 100
            final_score = (main_df["Final"][row] / 35) * 100

            total_score = assign_score + midterm_score + final_score
            ## Grade Calculate
        T_end = time.time()

        if i == 3 :
            run_df["Average"][index] = (run_df["1st Run"][index] + run_df["2nd Run"][index] + run_df["3rd Run"][index])/3
        else :
            run_df.iloc[index, i] = T_end - T_init
```

```python
cal_time(0, 50000)
cal_time(1, 100000)
cal_time(2, 150000)
cal_time(3, 200000)
cal_time(4, 250000)
run_df
```
